```python
df_categorical = df.loc[:,['cholesterol', 'smoke']]
sns.countplot(x="variable", hue="value",data= pd.melt(df_categorical));
```

```python
df_categorical = df.loc[:,[ 'alco', 'active']]
sns.countplot(x="variable", hue="value",data= pd.melt(df_categorical));
```

```python

correl = df.corr()
cmap = sns.diverging_palette(180, 12, as_cmap = True)

mask = np.zeros_like(correl, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize = (10, 14))

sns.heatmap(correl, mask = mask, cmap = cmap, vmax = 0.3, center = 0,annot = True, square = True, linewidths = 0.5, cbar_kws = {"shrink": .5});
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(x = 'disease', data = df, hue = 'smoke', palette = 'dark')
plt.show()
```
